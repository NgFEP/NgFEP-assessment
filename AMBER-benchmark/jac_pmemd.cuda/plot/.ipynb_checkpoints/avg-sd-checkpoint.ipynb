```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the list of file names
filenames = ['GTX_1080_Ti.dat','TITAN_V_case.dat','RTX_5000_case.dat','RTX_6000_cerebro.dat','GP100_cerebro.dat',
             'RTX_2080_Ti_cerebro.dat','RTX_3090_rioja.dat','RTX_5000_rioja.dat','RTX_A4500_rioja.dat']  # Add all your filenames here
#filenames = ['GTX_1080_Ti.dat','GP100_cerebro.dat',

device = ['GTX_1080_Ti','TITAN_V','RTX_5000','RTX_6000','GP100',
             'RTX_2080_Ti','RTX_3090','RTX_5000','RTX_A4500']
# Initialize lists to hold the averages and standard deviations
averages = []
standard_deviations = []

# Loop through each file, calculate the average and standard deviation of the second column
for file in filenames:
    # Check if file exists
    if not os.path.isfile(file):
        print(f"File {file} not found.")
        continue

    # Read the data from the file
    data = pd.read_csv(file, sep='\s+', header=None)  # Assuming whitespace delimiter, change if needed
    second_column = data.iloc[:, 1]  # Change index if the second column is not at position 1

    # Calculate average and standard deviation and append to lists
    averages.append(second_column.mean())
    standard_deviations.append(second_column.std())
# Create a bar plot with error bars
x_positions = np.arange(len(averages))
plt.bar(x_positions, averages, yerr=standard_deviations, align='center', alpha=0.7, ecolor='black', capsize=10)
print(averages)
# Add labels and title (optional)
#plt.xlabel('Files')
plt.ylabel('ns/day')
plt.title('JAC_results')
plt.xticks(x_positions, device, rotation='vertical')

# Display the plot
plt.tight_layout()
#plt.show()
plt.savefig('benchmarks_results.jpg', dpi=450, bbox_inches='tight')
```
