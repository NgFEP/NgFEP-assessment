```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A100_4', 'RTX_2080_8']
values = [6.50, 8.90]

# Create bar plot
plt.bar(categories, values, color='skyblue')

# Add title and labels
plt.title('1h1q~1h1r')
plt.xlabel('GPU')
plt.ylabel('ns/day')

# Show plot
#plt.show()
plt.savefig('benchmarks_1h1q~1h1r-AMBER-ACES_results.jpg', dpi=450, bbox_inches='tight')
```
